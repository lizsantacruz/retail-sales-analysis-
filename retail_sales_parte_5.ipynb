```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Proyecto I - Parte 5/retail_sales_dataset.csv')
```

```python
df.describe()
```

```python
#Utiliza histogramas y boxplots para visualizar la distribución de las variables numéricas clave.
plt.figure(figsize=(10, 5))
plt.hist(df['Total Amount'], bins=20, color='skyblue')
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.title("Histograma de Ventas")
plt.show()
```

```python
#Utiliza histogramas y boxplots para visualizar la distribución de las variables numéricas clave.
plt.figure(figsize=(10, 5))
plt.boxplot(df['Total Amount'], vert=False)
plt.xlabel("Ventas")
plt.title("Boxplot de Ventas")
plt.show()
```

```python
#cambio Date a formato fecha
df['Date'] = pd.to_datetime(df['Date'])

# Agrupacion por fecha y calculo de suma de ventas diarias
ventas_por_dia = df.groupby('Date')['Total Amount'].sum()

plt.figure(figsize=(12, 6))
plt.plot(ventas_por_dia.index, ventas_por_dia.values, color='blue')
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.title("Tendencia de Ventas por fechas")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Price per Unit'], df['Quantity'], alpha=0.5)
plt.xlabel("Precio")
plt.ylabel("Cantidad")
plt.title("Relación entre Precio y Cantidad vendida")
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 8), gridspec_kw={'height_ratios': [1, 4]})


ax[0].hist(df['Total Amount'], bins=20, color='skyblue')
ax[0].set_title("Histograma de Total de Ventas")
ax[0].set_ylabel("Frecuencia")

ax[1].boxplot(df['Total Amount'], vert=False)
ax[1].set_title("Boxplot de Total de Ventas")
ax[1].set_xlabel("Ventas")

plt.tight_layout()
plt.show()
```

```python
# Filtro columnas
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Relleno valores nulos
df_numeric = df_numeric.fillna(0)

# Correlacion con el metodo corr
correlation_matrix = df_numeric.corr()

# Creacion de el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Mapa de Calor de Correlación entre Variables")
plt.show()
```

```python
# Subplots para comparar distribuciones y relaciones entre variables
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Histograma de la primera variable
sns.histplot(df['Age'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribución de Edad')
axes[0, 0].grid(True)

# Histograma de la segunda variable
sns.histplot(df['Quantity'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribución de Cantidad')
axes[0, 1].grid(True)

# Gráfica de dispersión entre dos variables
sns.scatterplot(x='Age', y='Quantity', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Edad vs Cantidad')
axes[1, 0].grid(True)

# Gráfica de dispersión entre dos variables adicionales
sns.scatterplot(x='Total Amount', y='Age', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Total de ventas vs Edad')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
